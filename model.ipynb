# Model
This code can train or load a model, use it to make predictions and check the error.

```python
import main as mn
import paint as pt
```

## Train a Model

### Data Collect

```python
l,x,y=mn.get_transmissiondata1('data/data_transmission_2.dat')
pt.paint_spectra(l,x,y)
```

```python
x_train,x_validation,x_test,y_train,y_validation,y_test=mn.split_data(x,y)
```

### Training

```python
model,history=mn.train_model(x_train,x_validation,y_train,y_validation,optimizer='Adam',actfun='sigmoid',lr=0.00003,num_epochs=50000)
```

### Saving

```python
mn.save_model(model,history,l,x_train,x_validation,x_test,y_train,y_validation,y_test,route='',key='',)
```

## Load a Model

```python
model,history,l,x_train,x_validation,x_test,y_train,y_validation,y_test=mn.load_model(route='pre-trained/activation',key='sigmoid')
```

## Use a Model

```python
pt.paint_predictions(model,l,x_test,y_test)
```

```python
mn.error_model(model,x_train,x_validation,x_test,y_train,y_validation,y_test)
```
